# Matplotlib visualization

# How to Use matplotlib?

- Steps to create visualizations with matplotlib:
- 1- Choose the right chart type for representing your data.
- 2- Provide a title and labels for the plot axis.
- 3- Customize different aspects like size and colors.
- 4-  Plot your data.

```python
# import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
# 1-Line Chart
years = [1983, 1984, 1985, 1986, 1987]
total_populations = [8939007, 8954518, 8960387, 8956741, 8943721]
```

```python
years
```

```python
total_populations
```

```python
#To make a line plot with Matplotlib, we call plt.plot().
# plt.title(“My Title”) will add a title “My Title” to your plot
# plt.xlabel(“Year”) will add a label “Year” to your x-axis
# plt.ylabel(“Population”) will add a label “Population” to your y-axis

plt.plot(years, total_populations)
plt.title("Year vs Population in Bulgaria")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

```python
# 2-Scatter Plot
# To make a scatter plot with Matplotlib, we can use the plt.scatter() function.
temp = [30, 32, 33, 28.5, 35, 29, 29]
ice_creams_count = [100, 115, 115, 75, 125, 79, 89]
```

```python
plt.scatter(temp, ice_creams_count)
plt.title("Temperature vs. Sold ice creams")
plt.xlabel("Temperature")
plt.ylabel("Sold ice creams count")
plt.show()
```

```python
# 3-Histogram
# We divide the entire range of values into a series of intervals, and second, we count how many values fall into each interval.
# The intervals are also called bins.
# To make a histogram with Matplotlib, we can use the plt.hist() function. 
numbers = [0.1, 0.5, 1, 1.5, 2, 4, 5.5, 6, 8, 9]
```

```python
plt.hist(numbers, bins = 4)
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3-Box Plot
# plt.boxplot() used to plot Box Plot
values = [1, 2, 5, 6, 6, 7, 7, 8, 8, 8, 9, 10, 21]
```

```python
plt.boxplot(values)
plt.yticks(range(1, 22))
plt.ylabel("Value")
plt.show()
```

```python
# 4-Bar Chart- represents categorical data with rectangular bars.
# Each bar has a height corresponds to the value it represents.
# To make a bar chart with Maplotlib, we’ll need the plt.bar() function.

# Our data
labels = ["JavaScript", "Java", "Python", "C#"]
usage = [69.8, 45.3, 38.8, 34.4]
```

```python
# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))
```

```python
# Creating our bar plot
plt.bar(y_positions, usage,color="r")
plt.xticks(y_positions, labels)
plt.ylabel("Usage (%)")
plt.title("Programming language usage")
plt.show()
```

```python
# 5-Pie Chart-a circular plot, divided into slices to show numerical proportion. 
# To make a bar chart with Matplotlib, we can use the plt.pie() function. 
# The autopct parameter allows us to display the percentage value using the Python string formatting.

sizes = [25, 20, 45, 10]
labels = ["Cats", "Dogs", "Tigers", "Goats"]
```

```python
plt.pie(sizes, labels = labels, autopct = "%.2f")
plt.axes().set_aspect("equal")
plt.show()
```

```python
# 6-Quiver Plot
import numpy as np
n = 8
X, Y = np.mgrid[0:n, 0:n]
plt.quiver(X, Y)
```

```python
# 7-3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

```python
# 8-Violin Plot
n_datasets = 10
n_samples = 730
data = np.random.randn(n_datasets,n_samples)
```

```python
fig, axes = plt.subplots(1,3)
axes[0].violinplot([d for d in data])
axes[1].violinplot(data)
axes[2].violinplot(data.T)
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
```

```python
# make up some data in the interval [0 & 1]
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
```

```python
# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
```

```python
# log
plt.subplot(111)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
```

```python
# symmetric log
plt.subplot(111)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)
```

```python
# logit
plt.subplot(111)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
```
